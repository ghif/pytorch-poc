```python
import os
import json
import matplotlib.pyplot as plt
from plot_lib import plot_results, set_default
```

```python
# Constants
DATADIR = "/Users/mghifary/Work/Code/AI/data"
DATASET = "cifar10"
MODEL_DIR = "models"
```

```python
hist_path1 = os.path.join(MODEL_DIR, DATASET, f"plainnet18-randaug-exp2_hist.json")
hist_path2 = os.path.join(MODEL_DIR, DATASET, f"resnet18-randaug-exp2_hist.json")
```

```python
with open(hist_path1, "r") as f:
    hist1 = json.load(f)
print(hist1)
```

```python
with open(hist_path2, "r") as f:
    hist2 = json.load(f)
print(hist2)
```

```python
plot_results(
    hist1["train_accs"], 
    hist2["train_accs"], 
    xlabel="Epochs", title="CIFAR-10 Train Accuracy", 
    legend=["PlainNet18", "ResNet18"]
)
```

```python
plot_results(
    hist1["train_losses"], 
    hist2["train_losses"], 
    xlabel="Epochs", title="CIFAR-10 Train Loss", 
    legend=["PlainNet18", "ResNet18"]
)
```
